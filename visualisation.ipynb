```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Rentpredictionpreprocesseddata.csv')
df
```

```python
df = df.drop(['Unnamed: 0'], axis=1)
```

```python
df.info()
```

```python
df.describe()
```

```python
# selecting columns that are object
object_cols = df[['City', 'Names', 'location']]

# selecting columns that are float
float_cols = df.loc[:, ~df.columns.isin(['City', 'Names', 'location'])]
```

```python
fig, ax = plt.subplots(4, 3, figsize=(20, 20))

sns.histplot(data=float_cols, x='Bed', ax=ax[0][0])
sns.histplot(data=float_cols, x='Area', kde=True, ax=ax[0][1])
sns.histplot(data=float_cols, x='Rating', kde=True, ax=ax[0][2])
sns.histplot(data=float_cols, x='petsallowed', ax=ax[1][0])
sns.histplot(data=float_cols, x='pool', ax=ax[1][1])
sns.histplot(data=float_cols, x='fitnessCenter', ax=ax[1][2])
sns.histplot(data=float_cols, x='grill', ax=ax[2][0])
sns.histplot(data=float_cols, x='closet', ax=ax[2][1])
sns.histplot(data=float_cols, x='washerdryer', ax=ax[2][2])
sns.histplot(data=float_cols, x='countertops', ax=ax[3][0])
sns.histplot(data=float_cols, x='access', ax=ax[3][1])
sns.histplot(data=float_cols, x='Rent', kde=True, ax=ax[3][2])

plt.subplots_adjust(wspace=0.2)
plt.show()
```

```python
corrmat = float_cols.corr()

fig, ax = plt.subplots(figsize=(15, 8))
sns.heatmap(corrmat, ax=ax, annot=True)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 7))
sns.scatterplot(data=float_cols, x='Area', y='Rent', hue='Bed', ax=ax)
plt.title('Area vs rent of apartment classified by number of bedrooms')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 7))
sns.scatterplot(data=float_cols, x='Area', y='Rent', hue='fitnessCenter', ax=ax)
plt.title('Area vs rent of apartment classified by fitness center')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 7))
sns.scatterplot(data=float_cols, x='Area', y='Rent', hue='pool', ax=ax)
plt.title('Area vs rent of apartment classified by pool')
plt.show()
```

```python
fig, ax = plt.subplots(1, 4, figsize=(8, 5))

sns.boxplot(y=float_cols['Area'], ax=ax[0])
ax[0].set_xlabel('Area')
sns.boxplot(y=float_cols['Bed'], ax=ax[1])
ax[1].set_xlabel('bed')
sns.boxplot(y=float_cols['Rating'], ax=ax[2])
ax[2].set_xlabel('rating')
sns.boxplot(y=float_cols['Rent'], ax=ax[3])
ax[3].set_xlabel('Rent')

for subplot in ax:
    subplot.set_ylabel('')
plt.subplots_adjust(wspace=1.0)
plt.show()
```
