```python
def getdatas(city,i,url):
        print("inside",i)
        for i in range(1,i+1):
            scrapu=url+"/?"+"page="+str(i)
            print("scrapurl testing",scrapu)  
            webpage_response= requests.get(scrapu, headers=headers)
            time.sleep(5)
           
            
            #print(webpage_response.status_code)
            webpage=webpage_response.content
            soup = BeautifulSoup(webpage, 'html.parser')
            scrapurl=[]
            urls= soup.find_all("a", {"class": "property_link"})
            for a in urls:
                print ("Found the URL:", a['href'])
                scrapurl.append(a['href'])
            for u in scrapurl:
                webpage_response= requests.get(u, headers=headers)
                time.sleep(5)
                print(webpage_response.status_code)
                webpage=webpage_response.content
                soup = BeautifulSoup(webpage, 'html.parser')
                for rt in soup.find_all("section", {"id": "welcome"}):
                    for nm in rt.find_all("h1"):
                        print(nm.getText())
                        name.append(nm.getText())
                    
                print(name,len(name))
                l=[]
                for spd in soup.find_all("span", {"class": "specfic_details"}):
                    print(spd)
                    l.append(str(spd.getText()))
                if(len(l)!=3):
                            rent.append("?")
                            bed.append("?")
                            area.append("?")
                else:
                            bed.append(l[0].replace('BR', ''))
                            rent.append(l[1].replace('$', ''))
                            area.append(l[2].replace('sq ft', ''))
                r=[]
                for rt in soup.find_all("span", {"class": "star"}):
                    print(rt)
                    r.append(int(float(rt.getText())))
                if(len(r)==0):
                    rating.append("?")
                else:
                    rating.append(sum(r) / len(r))
                for add in soup.find_all("address"):
                    g=list(add.find_all("span"))
                    loc=""
                    for gi in g:
                        loc=loc+gi.getText()
                    print(loc)
                    location.append(loc.strip())
                pl=[]
                for pp in soup.find_all("section",{"id": "pets"}):
                    for kp in pp.find_all("li"):
                        pl.append(kp.getText())
                    print(pl)
                if(len(pl)==0):
                    petpolicy.append("?")
                else:
                    if("Dogs Allowed" in pl or "Cats Allowed" in pl):
                        petpolicy.append(1)
                    else:
                        petpolicy.append(0)
                print(petpolicy,len(petpolicy))
                    
                emlis=[]
                for amen in soup.find_all("section",{"id": "features_and_amen"}):
                    for emn in amen.find_all("li"):
                        emlis.append(emn.getText())
                if(len(emlis)==0):
                    pool.append("?")
                    gated.append("?")
                    fit.append("?")
                    access.append("?")
                    washerdryer.append("?")
                    countertops.append("?")
                    grill.append("?")
                    closet.append("?")
                    continue
                if "Walk In Closets" in emlis:
                    closet.append("1")
                else:
                    closet.append("0")
                        
                if "Grill" in emlis:
                    grill.append("1")
                else:
                    grill.append("0")
                            
                if "Granite Countertops" in emlis:
                    countertops.append("1")
                else:
                    countertops.append("0")
                            
                if "Washer/Dryer Hookup" in emlis:
                    washerdryer.append("1")
                else:
                    washerdryer.append("0")
                            
                if "Controlled Access" in emlis:
                    access.append("1")
                else:
                    access.append("0")
                            
                if "Pool" in emlis:
                    pool.append("1")
                else:
                    pool.append("0")
                if "Fitness Center" in emlis:
                    fit.append("1")
                else:
                    fit.append("0")
                        
                gated.append("?")
                bath.append("?")
                       
                                
                        
                
                
                  
                        
                    
                
                
```

```python
name=[]
bed=[]
area=[]
rent=[]
bath=[]
rating=[]
location=[]
petpolicy=[]
washerdryer=[]
countertops=[]
access=[]
pool=[]
fit=[]
gated=[]
grill=[]
closet=[]
getdatas("Denton",5,"https://www.apartmenthomeliving.com/denton-tx")
print(name,bed,area,rent,rating,location,petpolicy,washerdryer,countertops,access,pool,fit,gated,grill,closet)
print(len(name),len(bed),len(area),len(rent),len(rating),len(location),len(petpolicy),len(washerdryer),len(countertops),len(access),len(pool),len(fit),len(gated),len(grill),len(closet))
getdatas("Dallas",97,"https://www.apartmenthomeliving.com/dallas-tx")
print(name,bed,area,rent,rating,location,petpolicy,washerdryer,countertops,access,pool,fit,gated,grill,closet)
print(len(name),len(bed),len(area),len(rent),len(rating),len(location),len(petpolicy),len(washerdryer),len(countertops),len(access),len(pool),len(fit),len(gated),len(grill),len(closet))
```
